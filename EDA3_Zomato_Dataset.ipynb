```python
df_country = pd.read_excel("/content/zomato_dataset/Country-Code.xlsx")
df_country.head()
```

```python
df.columns
```

```python
# combining df and df_country
final_df = pd.merge(df, df_country, on='Country Code', how='left')
```

```python
final_df.head()
```

```python
# To check data types
final_df.dtypes
```

```python
final_df.Country.value_counts()
```

```python
country_name = final_df.Country.value_counts().index
country_name
```

```python
country_val = final_df.Country.value_counts().values
country_val
```

```python
# Pie chart
plt.pie(country_val, labels=country_name)
```

This seems very congested. Lets look at the top 3 countries.

```python
plt.pie(country_val[:3], labels=country_name[:3], autopct='%1.2f%%')
```

**Observation:**
<br>
Zomato maximum records or transactions are from India followed by United States and United Kingdom

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

**Observations:**
1. Rating between 4.5 - 4.9 --> Excellent
2. Rating between 4.0 - 4.4 --> Very Good
3. Rating between 3.5 - 3.9 --> Good
4. Rating betwenn 2.5 - 3.4 --> Average
5. Rating between 1.8 - 2.4 --> Poor

```python
ratings.head()
```

```python
import matplotlib

# increasing figure size
matplotlib.rcParams['figure.figsize'] = (12,6)
```

```python
sns.barplot(x="Aggregate rating", y="Rating Count", data=ratings, palette="deep")
```

```python
sns.barplot(x="Aggregate rating", y="Rating Count", data=ratings, hue="Rating color", palette=["white", 'red', 'orange', 'yellow', 'green', 'darkgreen'])
```

**Observation**
1. Not Rated count is very high
2. Maximum number of ratings are between 2.5 - 4.5


```python
# count plot
# plot of frequency of color in the ratings dataframe
sns.countplot(x="Rating color", data=ratings, palette=['blue', 'red', 'orange', 'yellow', 'green', 'darkgreen'])
```
