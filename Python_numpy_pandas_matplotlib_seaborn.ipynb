# Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.arange(0, 10)
y = np.arange(11,21)
```

```python
a = np.arange(40,50)
b = np.arange(50,60)
```

```python
# Scatter plot
plt.scatter(x,y,c='b')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title("Graph in 2D")

# saving the graph
plt.savefig('graph in 2d trial.png')
```

```python
y = x*x*x
```

```python
# plt plot
plt.plot(x,y, 'g*', linestyle='dashed', linewidth=2, markersize=12)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title("Graph in 2D")
```

```python
# Creating Subplots
plt.subplot(2,2,1)
plt.plot(x,x*x,'r')
plt.subplot(2,2,2)
plt.plot(x,x/2,'g')
plt.subplot(2,2,3)
plt.plot(x,x**(1/2),'b')
```

```python
np.pi
```

```python
# sine curve
x = np.arange(0,4*np.pi, 0.1)
y = np.sin(x)
plt.title("sine wave form")
plt.plot(x,y)
plt.show()
```

```python
print(x)
print(y)
```

```python
# sub plots
# sin and cos wave

x = np.arange(0, 5*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

fig, ax = plt.subplots(2, 1)
ax[0].plot(x, y_sin, 'r--')
ax[1].plot(x, y_cos)
plt.show()
```

```python
# bar plot
x = np.random.randint(10, size=10)
y = np.random.randint(10, size=10)
x2 = np.random.randint(10, size=10)
y2 = np.random.randint(10, size=10)
print(x,y,x2,y2, sep="\n")
plt.bar(x,y, color='g')
plt.bar(x2, y2, color='r')
plt.show()
```

```python
# Histogram
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,7])
plt.hist(a)
plt.title('Histogram')
plt.show()
```

```python
# box plot
data = [np.random.normal(0, std, 100) for std in range(1,5)]
plt.boxplot(data, vert=True, patch_artist=True)
plt.show()
```

```python
# Pie chart
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
explode = (0.1, 0, 0, 0)
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)
plt.axis('equal')
plt.show()
```

# Seaborn

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
# tip -> dependent feature
# others -> independent feature
```

```python
df.dtypes
```

```python
# Correlation with heatmaps
df[['total_bill', 'tip', 'size']].corr()
```

```python
sns.heatmap(df[['total_bill', 'tip', 'size']].corr())
```

Distribution plot
* joinplot
* distplot
* pairplot

```python
# Join Plot
sns.jointplot(x='total_bill', y='tip', data=df, kind='hex')
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind="reg")
```

```python
# Pair plot
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='sex')
```

```python
# Dist Plot
sns.distplot(df['total_bill'])
```

```python
sns.distplot(df['tip'],kde=False)
```

Categorical Plot
* boxplot
* voilinplot
* countplot
* bar plot

```python
df
```

```python
# Count plot
# need to specify only 1 axis, by default axis is x
sns.countplot(x='sex', data=df)
```

```python
sns.countplot(x='smoker', data=df)
```

```python
sns.countplot(y='day', data=df)
```

```python
# bar plot
# need to specify x and y axis

sns.barplot(x="total_bill", y="sex", data=df)
```

```python
# Box Plot

sns.boxplot(x='sex', y='total_bill', data=df)
```

```python
sns.boxplot(x='day', y='total_bill', data=df, palette='rainbow', hue="sex")
```

```python
sns.boxplot(data=df, orient="v")
```

```python
# violin plot
# helps to see both the distribution of data in terms of kernel density estimation and box plot

sns.violinplot(x="total_bill", y="day", data=df, palette="rainbow")
```
