## Feature Engineering Process

```python
# Handling Date_of_Journey categorical values
final_df['Date'] = final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month'] = final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year'] = final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head()
```

```python
# 2nd Technique by utilising lambda functons
# final_df['Date'] = final_df['Date_of_Journey'].apply(lambda x: x.split('/')[0])
```

```python
final_df.columns
```

```python
# converting date, month, year into integer
final_df['Date'] = final_df['Date'].astype(int)
final_df['Month'] = final_df['Month'].astype(int)
final_df['Year'] = final_df['Year'].astype(int)

final_df.info()
```

```python
# dropping Date_of_Journey
final_df.drop(['Date_of_Journey'], axis=1, inplace=True)
```

```python
final_df.columns
```

```python
# Handling Arrival_Time
final_df['Arrival_Time'].head(10)
```

```python
final_df['Arrival_Time'].str[:5]
```

```python
# final_df['Arrival_Time'] = final_df['Arrival_Time'].str[:5]
final_df['Arrival_Time'] = final_df['Arrival_Time'].apply(lambda x: x.split(' ')[0])
final_df.head(10)
```

```python
# splitting Arrival_Time into Arrival_hour and Arrival_min
final_df['Arrival_hour'] = final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arrival_min'] = final_df['Arrival_Time'].str.split(':').str[1]
```

```python
final_df.head()
```

```python
# converting object type to integer type
final_df['Arrival_hour'] = final_df['Arrival_hour'].astype(int)
final_df['Arrival_min'] = final_df['Arrival_min'].astype(int)

final_df.info()
```

```python
# dropping Arrival Time
final_df.drop(['Arrival_Time'], axis=1, inplace=True)
```

```python
final_df.head(10)
```

```python
# Handling Dep_Time categorical feature
final_df['Dep_Time'].unique()
```

```python
final_df_checkpoint = final_df.copy()
final_df_checkpoint
```

```python
# Splitting Dep_Time into Dep_hour and Dep_min
final_df['Dep_hour'] = final_df['Dep_Time'].apply(lambda x: x.split(":")[0])
final_df['Dep_min'] = final_df['Dep_Time'].apply(lambda x: x.split(':')[1])

final_df.head()
```

```python
# Converting object into integer type
final_df['Dep_hour'] = final_df['Dep_hour'].astype(int)
final_df['Dep_min'] = final_df['Dep_min'].astype(int)
```

```python
final_df.info()
```

```python
# Dropping Dep_Time
final_df.drop(['Dep_Time'], axis=1, inplace=True)
```

```python
final_df.head(3)
```

```python
final_df.info()
```

```python
# Handling Total_Stops
final_df['Total_Stops'].unique()
```

```python
final_df[final_df['Total_Stops'].isnull()]
```

```python
final_df['Total_Stops'] = final_df['Total_Stops'].map({
    'non-stop': 0,
    '1 stop': 1,
    '2 stops': 2,
    '3 stops': 3,
    '4 stops': 4,
    'nan': 1
})
```

```python
final_df.head()
```

**Observation**
<br>
As we have the number of stops and the source and destination feature, we can drop the route feature.

```python
# Dropping route
final_df.drop(['Route'], axis=1, inplace=True)
final_df.columns
```

```python
# Handling Duration categorical feature
final_df['Duration'].head()
```

```python
final_df['Duration'].unique()
```

```python
final_df[final_df['Duration']=='5m']
```

**Observation**
<br>
Mumbai to Hyderabad flight cannot have duration of 5 min with 2 stops in between. Henceforth, dropping these records.

```python
final_df.drop([6474, 2660], axis=0, inplace = True)
```

```python
final_df[final_df['Duration']=='5m']
```

```python
# splitting Duration into Duration_hour and Duration_min
final_df['Duration_hour'] = final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
final_df.head()
```

```python
final_df['Duration_hour'] = final_df['Duration_hour'].astype(int)
final_df.info()
```

```python
final_df['Duration'].str.split(' ').str[1]
```

```python
final_df['Duration_min'] = final_df['Duration'].str.split(' ').str[1]
final_df['Duration_min'].isnull().sum()
```

```python
final_df['Duration_min'] = final_df['Duration_min'].fillna('0m')
```

```python
final_df['Duration_min'].head(10)
```

```python
final_df['Duration_min'].str.split('m').str[0]
```

```python
final_df['Duration_min'] = final_df['Duration_min'].str.split('m').str[0]
final_df['Duration_min'] = final_df['Duration_min'].astype(int)
```

```python
final_df.info()
```

```python
# Dropping Duration
final_df.drop(['Duration'], axis=1, inplace=True)
final_df.info()
```

```python
# Using LabelEncoding
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
```

```python
final_df['Airline'] = labelencoder.fit_transform(final_df['Airline'])
final_df['Source'] = labelencoder.fit_transform(final_df['Source'])
final_df['Destination'] = labelencoder.fit_transform(final_df['Destination'])
final_df['Additional_Info'] = labelencoder.fit_transform(final_df['Additional_Info'])
```
