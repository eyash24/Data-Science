Data Cleaning

```python
sns.boxplot(x="Pclass", y="Age", data=train, palette="winter")
```

```python
train['Age'].describe()
```

```python
sns.jointplot(y="Age", x="Pclass", data=train, kind="kde")
```

```python
def impute_age(cols):
  Age = cols[0]
  Pclass = cols[1]

  if pd.isnull(Age):
    if Pclass == 1:
      return 37
    elif Pclass == 2:
      return 29
    elif Pclass == 3:
      return 24
  else:
    return Age
```

```python
train["Age"] = train[["Age", "Pclass"]].apply(impute_age, axis=1)
```

```python
sns.heatmap(train.isnull(), yticklabels=False, cmap="rainbow")
```

```python
# Dropping Cabin column due to large amount of null values
train.drop('Cabin', axis=1, inplace=True)
```

```python
train.head()
```

```python
train.isnull().sum()
```

```python
train.dropna(inplace = True)
```

```python
train.shape
```

Converting Categorical Features

```python
train.info()
```

```python
pd.get_dummies(train["Embarked"], drop_first=True, dtype=int).head()
```

```python
sex = pd.get_dummies(train["Sex"], drop_first=True, dtype=int)
embark = pd.get_dummies(train["Embarked"], drop_first=True, dtype=int)
```

```python
train.drop(["Sex", "Embarked", "Name", "Ticket"], axis=1, inplace=True)
```

```python
train.head()
```

```python
train = pd.concat([train, sex, embark], axis=1)
```

```python
train.head()
```

Building a Logistic Regression Model

```python
train.drop("Survived", axis=1).head()
```

```python
train.columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    train.drop("Survived", axis=1),
    train['Survived'], test_size=0.30,
    random_state=101
    )
```

```python
# Training and Predicting

from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train, y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
accuracy = confusion_matrix(y_test, predictions)
```

```python
accuracy
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, predictions)
accuracy
```
